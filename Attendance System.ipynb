```python
import os
import numpy as np
import face_recognition
import cv2
import pandas as pd
from datetime import datetime
```

```python
#creating list for images and storing names
path='train images'
images=[]
classNames=[]

myLst=os.listdir(path)
print(myLst)
```

```python
#loading the images in the list and names in classNames

for cl in myLst:
    curImg=cv2.imread(f'{path}/{cl}')
    images.append(curImg)
    classNames.append(os.path.splitext(cl)[0])

print(classNames)
    
```

```python
#Find the encodings

#create a function to find the encodings

def findEncodingd(images):
    #create an empty list to store the values of the enccodings
    encodeList=[]
    for img in images:
        img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
        encode=face_recognition.face_encodings(img)[0]
        encodeList.append(encode)
    
    return encodeList

#create a functio0n to store the attendance 

def markAttendance(name):
    with open('Attendance.csv','r+') as f:
        myDataList=f.readlines()
        nameList=[]
        for line in myDataList:
            entry=line.split(',')
            nameList.append(entry[0])
        if name not in nameList:
            now=datetime.now()
            dtString=now.strftime('%H:%M:%S')
            f.writelines(f'\n{name},{dtString}')

markAttendance('jack')
```

```python
#Getting the encoding values for the images

myEncode=findEncodingd(images)
print('Encoding Complete and number of items encoded are : ',len(myEncode))
```

```python
#Accessing the camera for recognising the face

cap=cv2.VideoCapture(0)

#Loop for Store the values

while True:
    
    #Reading the image from the video captured
    success,img=cap.read()
    
    #Set rhe Size and values for the images
    imgs=cv2.resize(img,(0,0),None,0.25,0.23)
    
    #Coverting the bgr image captured BGR to RGB
    imgs=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    
    
    #Finding the faces in the image
    faceCurFrame=face_recognition.face_locations(imgs)
    encodeCurFrame=face_recognition.face_encodings(imgs,faceCurFrame)
    
    #comparing the values with the stored list
    
    for encodeFace,faceLoc in zip(encodeCurFrame,faceCurFrame):
        result=face_recognition.compare_faces(myEncode,encodeFace)
        faceDistance=face_recognition.face_distance(myEncode,encodeFace)
        
        #Prints the encoding face distance
        print(faceDistance)
        
        matchIndex=np.argmin(faceDistance)
        
        if result[matchIndex]:
            name=classNames[matchIndex].upper()
            print(name)
            
            y1,x2,y2,x1=faceLoc
            
            #increase the size of the image as we reduced earlier
            y1,x2,y2,x1=y1*4,x2*4,y2*4,x1*4
            
            cv2.rectangle(img,(x1,y1),(x2,y2),(0,255,0),2)
            cv2.rectangle(img,(x1,y2-35),(x2,y2),(0,255,0),cv2.FILLED)
            cv2.putText(img,name,(x1+6,y2-6),cv2.FONT_HERSHEY_COMPLEX,1,(255,255,255),2)
            markAttendance(name)
    cv2.imshow('Webcam',img)
    cv2.waitKey(1)
```
